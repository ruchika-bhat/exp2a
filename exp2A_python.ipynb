```python
# elements of the list can repeat
listfruits=["apple","orange","apple","cherry"]
print(listfruits)
```

```python
# finding the length of list
list1=[2,2.5,True,"ruchika"]
print(list1)
print(len(list1))
```

```python
# adding new elements to the list later
listfruits=["apple","orange","cherry"]
listfruits.append("banana")
print(listfruits)
```

```python
# printing only one element
listfruits=["apple","orange","cherry"]
print(listfruits[0])
```

```python
# printing two elements only
listfruits=["apple","orange","cherry"]
print(listfruits[0],listfruits[-2])
```

```python
listfruits=["apple","orange","cherry"]
print(listfruits[0])
print(listfruits[-2])
listfruits.append("banana")
print(listfruits[0])
print(listfruits[-2])
print(listfruits)
```

```python
# adding new element to the list at a certain position
listfruits=["apple","orange","cherry"]
listfruits.insert(1,"banana")
print(listfruits)
# if integer writing without double qoutes
```

```python
# replacing element with another element
listfruits=["apple","orange","cherry"]
listfruits[1]="banana"
print(listfruits)
```

```python
# combining two different lists together(using addition)
list1=["apple","orange","cherry"]
list2=["mango","pineapple","banana"]
list=(list1+list2)
print(list)

# combining two different lists together(using extend command)
list1=["apple","orange","cherry"]
list2=["mango","pineapple","banana"]
list1.extend(list2)
print(list1)
```

```python
# printing elements point-to-point wise
list1=["apple","banana","cherry","kiwi","mango","pineapple"]
print(list1[2:5])
print(list1[3:6])
print(list1[1:]) #prints from 1 element till end
print(list1[:4]) #prints from starting to 3 element
```
